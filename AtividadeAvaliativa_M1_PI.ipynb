```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

# img = cv2.imread('/images/lena_original.png', 0)
# img = cv2.imread('/images/jetplane_original.tif', 0)
img = cv2.imread('/images/wallpaper_original.jpg', 0)

media = 0
desvio = 25
sigma = desvio * 0.5

gaussian = np.random.normal(media, sigma, (512, 512))

imgNormalize = cv2.normalize(gaussian, img, 0, 255, cv2.NORM_MINMAX, dtype=-1)

imgGaussiana = cv2.add(img, imgNormalize.astype("uint8"))

cv2_imshow(img)
cv2_imshow(imgGaussiana)
```

```python
from math import log10, sqrt
import cv2
import numpy as np

def PSNR(original, processada):
	mse = np.mean((original - processada) ** 2)
	if(mse == 0): # Se o mse for 0, quer dizer que no tem rudo, logo o PSNR  100.
		return 100
	max_pixel = 255.0
	psnr = 20 * log10(max_pixel / sqrt(mse))
	return psnr

# original = cv2.imread("/images/lena_original.png")
# processada = cv2.imread("/images/lena_processada.png", 1)

# original = cv2.imread("/images/jetplane_original.tif")
# processada = cv2.imread("/images/jetplane_processada.png", 1)

original = cv2.imread("/images/wallpaper_original.jpg")
processada = cv2.imread("/images/wallpaper_processada.png", 1)
psnrValue = PSNR(original, processada)
print(f"PSNR: {psnrValue} dB")
```
